#### Import Library and Data

```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url ="http://www.eee.ntu.edu.sg/programmes/CurrentStudents/undergraduate/undergraduatefull-time/Pages/CoursesCatalogue.aspx"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text,"html.parser")
```

After inspect the html element, it's found that all data are inside 'td' tags

```python
list_course = soup.find_all('td')
final = [txt.find_all(text=True) for txt in list_course]
```

```python
len(final)
```

#### Normalize data

```python
core_data=final[4:len(final)-4]
core_data[13]=core_data[1]
core_data
```

```python
for i in range (len(core_data)):
    if core_data[i]==['\u200b']:
        core_data[i]=['']
        
for i in range (len(core_data)):
    if core_data[i]==['']:
        core_data[i]=[]
core_data
```

```python
core_data_copy=core_data.copy()
```

```python
for i in range (len(core_data_copy)):
    if '\u200b' in core_data_copy[i]:
        core_data_copy[i].pop(core_data_copy[i].index('\u200b'))

for i in range (len(core_data_copy)):
    if ' ' in core_data_copy[i]:
        core_data_copy[i].pop(core_data_copy[i].index(' '))
core_data_copy
```

```python
core_value=[]
for i in core_data_copy:
    try:
        core_value.append(i[0])
    except:
        core_value.append('')
```

```python
core_value
```

#### Convert to Excel file

```python
value=np.array(core_value)
value=value.reshape(len(value)//6,6)
value
```

```python
df=pd.DataFrame(value)
df
```

```python
df.columns=['Course Code', 'Name', 'Course Type', 'AU', 'Sem 1', 'Sem 2']
df.set_index(pd.Index(list(range(1,110))))
df
```

```python
df.to_excel("EEE NTU Courses.xlsx")
```
